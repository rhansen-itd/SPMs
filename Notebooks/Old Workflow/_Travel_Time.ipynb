**Import Dependencies**

```python
import pandas as pd
import os
import datetime as dt
#from sklearn.linear_model import LinearRegression
import spmfunctions.spmfunctions as spm
```

**INRIX Travel Time**

```python
intersections=['Chinden',#'Colchester',
               'Fairview','Franklin',
               'HobbleCreek',#'I84EB','I84WB','IslandWoods','McMillan','Pine','RiverValley',
               'StLuke','Ustick',
               'Riverside','Wainwright']

atr_key={'Riverside':{4:('270','Negative'),8:('270','Positive')},
     'Chinden':{2:('337','Negative'),4:('330','Negative'),6:('337','Positive'),8:('330','Positive')},
     'HobbleCreek':{4:('330','Negative'),8:('330','Positive')}}
atr_key={}

b_l=60
wtf=None #'xl','csv'
xl_mode='w' #write='w', append='a'
sfx='_4-12'
phases=[2,4,6,8]

resdict={}
for intx in intersections:    
    print('Starting {} {}'.format(intx,dt.datetime.now()))
    if intx in list(atr_key.keys()):
        atr=atr_key[intx]
    else:
        atr=None
    
    try:
        rd=spm.combine_aog_sf_tt(intx,bin_len=b_l,sfx=sfx,write_to_file=wtf,xl_mode=xl_mode,atr=atr)
        resdict[intx]=rd
    except:
        print('Error processing {}'.format(intx))
    print('Finished {} {}'.format(intx,dt.datetime.now()))

            
```

```python
t_ba='7-20-2023'
sfx='_SF_AOG'

anova_rd={}
for intx, r_d in resdict.items():
    #if True:
    try:
        anova_rd[intx]=spm.tt_vol_anova(r_d, t_ba,atr_vol=True)
        anova_rd[intx].to_csv('./ACHD_Eagle-{}/{}_ANOVA{}.csv'.format(intx,intx,sfx),index=False)
    except:
        print('Error processing {}'.format(intx))
```

```python
resdf=pd.DataFrame()

for intx in intersections:
    csvs=os.listdir('./ACHD_Eagle-{}'.format(intx))
    for f in [x for x in csvs if 'ANOVA_SF_AOG' in x]:
        if 'single_adv' in f:
            det=2
        else:
            det=1
        df=pd.read_csv('./ACHD_Eagle-{}/{}'.format(intx,f))
        df.loc[:,'Int']=intx
        df.loc[:,'AdvDet']=det
        resdf=pd.concat([resdf,df])
```

```python
resdf.to_csv('ANOVA_results.csv')
```
