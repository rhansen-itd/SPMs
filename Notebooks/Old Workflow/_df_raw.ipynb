```python
#Combines csv files from Miovision output into compiled csv files for import with read_df_raw
int_key={'Colchester':392,'Fairview':213,'Franklin':270,'HobbleCreek':361,'I84EB':339,#'Chinden':250,
 'I84WB':269,'IslandWoods':341,'McMillan':272,'Pine':323,'Riverside':325,'RiverValley':423,
 'StLuke':201,'Ustick':271,'Village':480,'Wainwright':430}

f_comp='compiled_2310.csv'
path_csv_r='./Downloaded_CSVs/'
same_folder=False

date_codes=['20231001','20231016']


for intx in int_key.keys():
#if True:
    int_code=int_key[intx]
    path_csv_w='./ACHD_Eagle-%s/CSV'%intx
    if same_folder:
        path_csv_r=path_csv_w

    #path_csv='./2nd & american legion/CSV'
    #f_comp='compiled.csv'
    
    df_raw1=pd.DataFrame()
    l=0
    
    if same_folder:
        f_list=[x for x in os.listdir(path_csv_r) if (x[-3:]=='csv')]
    else:
        f_list=[x for x in os.listdir(path_csv_r) if (x[-17:-9] in(date_codes) and x[0:3]==str(int_code))]
    
    for f_csv in f_list:
        print(f_csv)
        f_path=os.path.join(path_csv_r, f_csv)
        
        with open(f_path) as f:
            l1=f.readline()
            if l1.startswith('Signal,'): #Miovision
                skiprows=5
                header=None
                usecols=[0,1,2,3]
                names=['TS_start', 'Code', 'Desc','ID']
            elif l1.startswith('Timestamp'): #Econolite
                skiprows=8
                header=None
                usecols=[0,1,2]
                names=['TS_start', 'Code', 'ID']
            else: #previously processed (no header)
                skiprows=None
                header=None
                if 'ACHD' in path_csv_r:
                    usecols=[0,1,2,3]
                    names=['TS_start', 'Code', 'Desc','ID']
                else:
                    usecols=[0,1,2]
                    names=['TS_start', 'Code', 'ID']
            
#            if f_csv=='compiled6-10_23.csv':
#                nrows=10000
#            else:
            nrows=None
                
            
            df_rd=pd.read_csv(f_path, usecols=usecols, skiprows=skiprows, names=names, header=header,nrows=nrows)

        df_raw1=pd.concat([df_raw1, df_rd])

    df_raw1.to_csv(os.path.join(path_csv_w, f_comp),index=False, header=False)
```
