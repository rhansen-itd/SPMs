```python
import pandas as pd
import os
import datetime as dt
import spmfunctions.spmfunctions as spm
import warnings
warnings.filterwarnings('ignore')
```

```python
intx='SH-44-Hartley'
f_pkl='df_raw.pk'
yymm=''
sfx_sum='_hourly'

df_mvmt,df_exc,df_adv,df_sb=spm.det_cfg('./{}/Configuration/'.format(intx))
ph_dets=[(int(i[1]),int(j.loc['Det 1'])) for i,j in df_sb.iterrows()]

path_pkl = '.\\{}\df_raw'.format(intx)

df_raw = pd.read_pickle(os.path.join(path_pkl,f_pkl))

df_res=pd.DataFrame()

for ph, det in ph_dets:
    print('Starting Ph {} {}'.format(ph, dt.datetime.now()))
    if len(df_raw[(df_raw.ID==ph)&(df_raw.Code==1)])>1:
        df_res=df_res.append(spm.split_failures(df_raw, ph, det))
    print('Ph {} Complete {}'.format(ph, dt.datetime.now()))

if not os.path.exists('./{}/Split Failures'.format(intx)):
    os.mkdir('./{}/Split Failures'.format(intx))

df_res.sort_values(['Phase','Time']).to_csv('./{}/Split Failures/{}_Split_Failures{}.csv'.format(intx,intx,yymm))#index=False

spm.bin_SF(df_res).to_csv('./{}/Split Failures/{}_Split_Failures{}{}.csv'.format(intx,intx,yymm,sfx_sum), index=False)
```
