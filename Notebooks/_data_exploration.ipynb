```python
main_df = df.copy()
detection_df = cfg_dict['Arrivals'].copy()

# We are going to change count detectors to be presence. Change this function to either use only code 81 or use both 81 and 82 (but only when both are during red)

def process_dataframe(main_df: pd.DataFrame, detection_df: pd.DataFrame) -> pd.DataFrame:
    # Extract only even-numbered phase columns early
    phase_columns = [col for col in main_df.columns if col.startswith('Ph ') and 'Status' in col]
    even_phases = {int(col.split()[1]) for col in phase_columns if int(col.split()[1]) % 2 == 0}
    
    # Keep only relevant columns in main dataframe
    main_df = main_df[['TS_start', 'Cycle_start', 'Code', 'ID', 't_cs', 'Duration'] + 
                      [f'Ph {phase} Status' for phase in even_phases]]
    
    # Filter main dataframe for relevant codes
    main_df = main_df[main_df['Code'].isin([82, 10])].copy()
    
    # Extract only even-numbered phase stop bar columns in detection dataframe
    stop_bar_columns = [col for col in detection_df.columns if col.startswith('P') and 'Stop Bar' in col]
    even_stop_bars = {int(col[1:].split()[0]) for col in stop_bar_columns if int(col[1:].split()[0]) % 2 == 0}
    even_phases &= even_stop_bars
    
    detection_df = detection_df[[f'P{phase} Stop Bar' for phase in even_phases]]
    
    # Convert detection dataframe stop bars to lists of integers and map detectors to phases
    detector_phase_map = {}
    for phase in even_phases:
        col_name = f'P{phase} Stop Bar'
        detection_df[col_name] = detection_df[col_name].astype(str).apply(lambda x: [int(i) for i in x.split(',') if i.isdigit()])
        for det in detection_df[col_name].explode().dropna().astype(int):
            detector_phase_map[det] = phase
    
    # Filter main dataframe to only include relevant detectors and even phases
    main_df = main_df[((main_df['Code'] == 10) & (main_df['ID'] % 2 == 0)) | ((main_df['Code'] == 82) & (main_df['ID'].isin(detector_phase_map)))]
    
    # Merge main and detection dataframes to align timestamps
    merged_df = main_df.merge(detection_df, left_on='TS_start', right_index=True, how='outer').sort_values('TS_start')
    
    # Forward fill detector columns and remove rows added from detection_df
    for phase in even_phases:
        col_name = f'P{phase} Stop Bar'
        merged_df[col_name] = merged_df[col_name].ffill()
    merged_df = merged_df[~merged_df['Code'].isna()]
    
    # Convert Code back to integer
    merged_df['Code'] = merged_df['Code'].astype(int)
    
    # Split data into phase events (Code==10) and detector events (Code==82)
    phase_events = merged_df[merged_df['Code'] == 10].copy()
    detector_events = merged_df[merged_df['Code'] == 82].copy()
    
    # Rename ID to Phase in phase_events for merging
    phase_events = phase_events.rename(columns={'ID': 'Phase'})
    phase_events['TS_red'] = phase_events['TS_start']
    
    # Assign corresponding phase to each detector event
    detector_events['Phase'] = detector_events['ID'].map(detector_phase_map)
    
    # Drop detection event rows where the corresponding phase status is not 'R' or 'Rc'
    for phase in even_phases:
        phase_status_col = f'Ph {phase} Status'
        detector_events = detector_events[~((detector_events['Phase'] == phase) & ~detector_events[phase_status_col].isin(['R', 'Rc']))]
    
    # Merge phase events with detector events on Phase using an outer join to keep times aligned
    merged_events = detector_events.merge(phase_events[['TS_start', 'Phase', 'TS_red']], 
                                          on=['TS_start','Phase'], 
                                          how='outer', 
                                          suffixes=('', '_red'))
    
    # Sort and forward-fill TS_red within each phase group
    merged_events = merged_events.sort_values(['Phase', 'TS_start'])
    merged_events['TS_red'] = merged_events.groupby('Phase')['TS_red'].ffill()
    
    # Compute time difference (t_r) between detector event and preceding phase red event
    merged_events['t_r'] = (merged_events['TS_start'] - merged_events['TS_red']).dt.total_seconds()
    
    # Select relevant columns and return final dataframe
    result_df = merged_events[['TS_start', 't_r', 'ID', 'Phase', 'Duration']]

    return result_df.dropna()
```
