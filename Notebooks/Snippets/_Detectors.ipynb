```python
import pandas as pd
import os
import datetime as dt
import spmfunctions.spmfunctions as spm
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_rows', 500)
```

```python
intx='ACHD_Chinden-SH16'
sfx='2405'

df_raw=pd.read_pickle('./{}/df_raw/df_raw{}.pk'.format(intx,sfx))
```

```python
df_comb=spm.comb_gyr_det(df_raw)
```

```python
ph=4#8
dets=[59,60,61]#[51,52,53]

df_dt=df_comb[((df_comb.ID.isin(dets))&(df_comb.Code.isin([82])))|((df_comb.Code==1)&(df_comb.ID==ph))]
df_dt
```

```python
df_res=pd.DataFrame()

for t, gb in df_dt.groupby('Cycle_start'):
    
    gb=gb.sort_values(['Code','TS_start']).reset_index(drop=True)
    if gb.iloc[0,1]!=1:
        continue
    gb=gb[(gb.Code==82)]
    gb.loc[:,'Delta_Time']=gb.t_cs-gb.t_cs.shift()
    gb=gb[(gb.ID==gb.ID.shift()-1)]
    
    df_res=pd.concat([df_res,gb])
```
