```python
def rate_dfs(df_res, res_dict,pct=1):

    df_res.index.name='TS'
    #df_mode=df_res[(df_res.g>0.9*df_res.g.mode()[0])&(df_res.g<1.1*df_res.g.mode()[0])]
    df_mode=df_res[(df_res.split>0.9*df_res.split.mode()[0])&(df_res.split<1.1*df_res.split.mode()[0])]
    q_srs=df_mode.groupby('TS').Q.sum()
    df_pct=df_mode.loc[q_srs[q_srs>=q_srs.quantile((100-pct)/100)].index]

    df_rate=pd.DataFrame()
    df_inst=pd.DataFrame()
    for det in dets:
        df_det=pd.DataFrame()
        df_det_i=pd.DataFrame()
        for key in df_pct.index.unique():
            df=res_dict[f'{det} {key}'][res_dict[f'{det} {key}'].ID==det]
            df.loc[:,'t']=df.t.apply(lambda x: round(2*x,0)/2)
            df_det=pd.concat([df_det,df.set_index('t').cumulative_rate.rename(f'{det} {key}')],axis=1)
            df_det_i=pd.concat([df_det_i,df.set_index('t').inst_rate.rename(f'{det} {key}')],axis=1)
        df_det.reindex([x/2 for x in list(range(int(2*df_det.index.min()),int(2*df_det.index.max())+1))])
        df_det.loc[:,f'{det} Mean']=df_det.where(df_det.bfill().isna(),df_det.interpolate()).dropna(thresh=5).mean(axis=1)
        df_rate=pd.concat([df_rate,df_det],axis=1)
        df_det_i.reindex([x/2 for x in list(range(int(2*df_det_i.index.min()),int(2*df_det_i.index.max())+1))])
        df_det_i.loc[:,f'{det} Mean']=df_det_i.where(df_det_i.bfill().isna(),df_det_i.interpolate()).dropna(thresh=5).mean(axis=1)
        df_inst=pd.concat([df_inst,df_det_i],axis=1)
    #df_rate = df_rate.where(df_rate.bfill().isna(), df_rate.ffill().fillna(0))
    #df_rate = df_rate.where(df_rate.bfill().isna(), df_rate.interpolate())
    cols=[c for c in df_rate.columns if 'Mean' not in c]
    df_rate.loc[:,'Mean']=df_rate.loc[:,cols].interpolate().mean(axis=1)
    df_inst.loc[:,'Mean']=df_inst.loc[:,cols].interpolate().mean(axis=1)
    
    return df_pct, df_rate, df_inst



def flow_rate(df,ph,dets,max_lost=10):

    #df_gy=df[(df.ID==ph)&(df.Code.isin([1,8,11]))]
    df_gy=df[(df.ID==ph)&(df.Code.isin([1,11]))]
    df_gy.loc[:,'TS_end']=df_gy.TS_start.shift(-1)
    #df_gy.loc[:,'ryc']=(df_gy.TS_end-df_gy.TS_start).shift(-1).dt.total_seconds()
    df_gy=df_gy.loc[(df_gy.Code==1),['TS_start','TS_end',#'ryc',
                                     'Coord_plan']].dropna()
    #df_gy=df_gy[df_gy.ryc<rycmax]

   
    df_res=pd.DataFrame()
    res_dict={}
            
    for det in dets:
        df_det=df[(df.Code==81)&(df.ID==det)]
        for _,row in df_gy.iterrows():
            srs=pd.Series(index=['Det','Coord_plan','t_max','max_rate','cumulative','split','lost',#'g',
                                 'Q'])
            srs['Det']=det
            df_cyc=filter_date(df_det,row.TS_start,row.TS_end,'TS_start',False)
            if len(df_cyc)==0:
                continue
            
            r_lost=(row.TS_end-df_cyc.TS_start.iloc[-1]).total_seconds()
            if r_lost > max_lost:
                continue
            
            df_cyc.reset_index(inplace=True,drop=True)
            df_cyc.loc[:,'cumulative']=df_cyc.index+1
            df_cyc.loc[:,'t']=(df_cyc.TS_start-row.TS_start).dt.total_seconds()
            df_cyc.loc[:,'cumulative_rate']=(3600*df_cyc.cumulative)/((df_cyc.t+r_lost))
            #df_cyc.loc[:,'cumulative_rate']=(3600*df_cyc.cumulative)/((df_cyc.t+row.ryc))
            
            for _, gb in df_cyc.groupby('ID',as_index=False):
                df_cyc.loc[gb.index,'headway']=(gb.TS_start.shift(-1)-gb.TS_start).dt.total_seconds()
            df_cyc.loc[:,'inst_rate']=3600/df_cyc.headway
            t_max,rate,cumulative=df_cyc[df_cyc.cumulative_rate==df_cyc.cumulative_rate.max()].loc[:,['t','cumulative_rate','cumulative']].values[0]
            #g_len=(row.TS_end-row.TS_start).total_seconds()
            split=(row.TS_end-row.TS_start).total_seconds()
            Q=df_cyc.cumulative.max()
            cp=row.Coord_plan
            srs[['Coord_plan','t_max','max_rate','cumulative','split','lost',#'g',
                 'Q']]=[cp,t_max,rate,cumulative,split,r_lost,#g_len,
                        Q]
            srs.name=row.TS_start
            df_res = pd.concat([df_res, srs.to_frame().T], ignore_index=False)
            res_dict[f'{det} {row.TS_start}']=df_cyc
    
    return df_res,res_dict



def plot_flow_plotly(df_rate, df_inst, save_path, intx, ph, rolling=1):
    fig = make_subplots(rows=1, cols=1, subplot_titles=[f"{intx} Ph{ph} Flow Rate"])
    color={'Mean':'gray'}
    #symbol={}
    colors=['red', 'blue', 'green', 'magenta']
    #symbols=['circle','square','diamond','triangle-up']
    dets = list(set([x.split(' ')[0] for x in df_rate.columns[:-1]]))

    for i, d in enumerate(dets):
        color[d] = colors[i]
        #symbol[d] = symbols[i]

    for c in df_rate.columns:
        df_plt = df_rate.loc[:, c].dropna().sort_index()
        det = c.split(' ')[0]

        if 'Mean' in c:
            fig.add_trace(go.Scatter(
                x=df_plt.index,
                y=df_plt.values,
                mode='lines',
                line=dict(color=color[det]),
                hoverinfo='x+y+text',  # Show x, y, and text (column name)
                text=[f'{c}: {val}' for val in df_plt.values]  # Text to display on hover
            ))
            df_plt2 = df_inst.loc[:, c].sort_index().transform(lambda x: x.rolling(rolling,center=True).mean()).dropna()
            fig.add_trace(go.Scatter(
                x=df_plt2.index,
                y=-df_plt2.values,
                mode='lines',
                line=dict(color=color[det], dash='dash'),
                hoverinfo='x+y+text',  # Show x, y, and text (column name)
                text=[f'{c}: {val}' for val in df_plt2.values]  # Text to display on hover
            ))
        else:
            fig.add_trace(go.Scatter(
                x=df_plt.index,
                y=df_plt.values,
                mode='markers',
                marker=dict(opacity=0.5, color=color[det]),#symbol=symbol[det], 
                hoverinfo='x+y+text',  # Show x, y, and text (column name)
                text=[f'{c}: {val}' for val in df_plt.values]  # Text to display on hover
            ))

    fig.update_layout(
        xaxis=dict(title='Time (s)', type='linear'),
        yaxis=dict(title='Flow Rate (vphpl)', type='linear'),
        showlegend=False
    )

    fig.write_html(save_path)
```
